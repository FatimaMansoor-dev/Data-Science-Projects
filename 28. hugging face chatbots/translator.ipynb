```python
from transformers import pipeline
import torch
```

```python
translator = pipeline(task='translation',
                      model = 'facebook/nllb-200-distilled-600M',
                      torch_dtype = torch.bfloat16)
```

```python
text = 'hi my name is fama'
```

```python
trans = translator(text,
                   src_lang = 'eng_Latn',
                   tgt_lang = 'fra_Latn')
```

```python
trans
```

```python
trans = translator('comment vas-tu?',
                   src_lang = 'fra_Latn',
                   tgt_lang = 'eng_Latn')
```

```python
trans
```
